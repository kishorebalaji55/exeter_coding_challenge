## Using string replace function for replacing the English words with French words

```python
def multipleReplace(text, wordDict):
    check=[]
    frequency = []
    for key in wordDict:
        frequency.append([a.start() for a in re.finditer(key, text)])
        text = text.replace(key, wordDict[key])
        check.append([a.start() for a in re.finditer(wordDict[key], text)])
    return text,frequency,check
```

```python
str1,frequency,check = multipleReplace(data, dic)
```

The following lines of codes are used to understand how many words have been replaced in the text input and the comparision of the number of occurance of the english words, before replacement, and french words, after the replacement. The number of words that have been replaced are 878, which will be checked for sanity in the following tabs.

```python
count=0
for i in range(len(frequency)):
    print(len(frequency[i]),'\t',len(check[i]))
    if len(frequency[i])>0:
        count+=1 
print("Number of words with replacements are : ", count)
```

This tab works on counting the number of words in the list of lists to calculate the frequency of the occurance of each word in the text input file.

```python
freq = []
for i in range(len(frequency)):
    freq.append(len(frequency[i]))
#freq
```

## Creating a dataframe and then a CSV file for the frequency data

```python
Dict=[{'English word':eng, 'French word':fre, 'Frequency':fr} for eng,fre,fr in zip(eng_list,fre_list,freq)]
```

```python
df = pd.DataFrame (Dict, columns = ['English word','French word','Frequency'])
df
```

```python
df.to_csv("frequency.csv", index=None)
```

### Finding the Unique list of words that was replaced with French words from the dictionary

```python
list_of_words = []
for i in range(len(df)):
    if df['Frequency'][i]>0:
        list_of_words.append(df['English word'][i])
#list_of_words
```
